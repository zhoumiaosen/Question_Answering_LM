```python
import matplotlib.pyplot as plt
```

```python
#RoBERTa Based Model
results = {'exact_match': 0.9460737937559129, 'f1': 0.9293147722665225}
# Data
categories = ["Exact Match", "F1 Score"]
scores = [results["exact_match"]*100, results["f1"]*100]

# Plot
plt.figure(figsize=(6, 4))
plt.bar(categories, scores, color=['blue', 'green'])
plt.title("RoBERTa Based Model Performance on SQuAD Validation Set", fontsize=14)
plt.ylabel("Score (%)", fontsize=12)
for i, score in enumerate(scores):
    plt.text(i, score + 0.5, f"{score:.2f}%", ha='center', fontsize=12)
plt.ylim(0, 100)
plt.show()
```
